# Problem 2

*Solve the following ODE using the explicit Euler, implicit Euler, and modified midpoint methods. Use a stepsize of dt=0.2, 0.4, 0.8, and 1.6. Compare the results graphically with the exact solution. Make four plots (one for each stepsize, use the subplot command if you like), and plot Exact, EE, IE, MM on each plot. A good x and y range are 0-8 and 1-4, respectively.
$$\begin{align}&\\ \frac{dy}{dt} &= ye^{-t}, \\
y(t=0) &= 1.\end{align}$$*

```python
def p2_rhsf(y,t) :
    return y*np.exp(-t)
```

```python
def Integrate(y0, frhs, Tend, dt, method) :
    """ 
    Solve an ode problem dy/dt = frhs(y,t) from t=0 to t=Tend using stepsize dt.
    y0 is the initial condition.
    Set method = one of "EE", "IE", "ME" 
    returns y(t),t
    """
    
    nsteps = int(Tend/dt)
    ntimes = nsteps + 1
    t = np.linspace(0.0,Tend,ntimes)
    y = np.zeros(ntimes)
    y[0] = y0
        
    for n in range(nsteps) :
        if method == "EE" :
            y[n+1] = y[n] + dt*frhs(y[n],t[n])
        elif method == "IE" :
            y[n+1] = y[n] / (1-dt*np.exp(-t[n+1]))        # note, this one is hard-coded.
        elif method == "MM" :
            yhalf = y[n] + 0.5*dt * frhs(y[n],t[n])
            y[n+1] = y[n] + dt*frhs(yhalf,t[n]+0.5*dt)
        else :
            print "ERROR, method must equal one of: EE, IE, MM"
            return y,t
    
    return y,t
```

```python
dt = np.array([0.2,0.4,0.8,1.6])                # array of timestep sizes
y0 = 1.0                                        # initial condition
Tend = 8.0                                      # integration end time
texact = np.linspace(0.0,Tend,1000)             # domain for exact solution
yexact = y0*np.exp(1.0-np.exp(-texact))         # exact solution

for idt in range(4) :                           # make four plots for 4 dt's: dt[idt]
    yEE,tEE = Integrate(y0, p2_rhsf, Tend, dt[idt], "EE")      
    yIE,tIE = Integrate(y0, p2_rhsf, Tend, dt[idt], "IE")
    yMM,tMM = Integrate(y0, p2_rhsf, Tend, dt[idt], "MM")
    #plt.figure(idt)
    plt.subplot(2,2,idt+1)
    plt.plot(texact,yexact,'k-',label='exact')
    plt.plot(tEE,yEE,'b-',label='EE')
    plt.plot(tIE,yIE,'g-',label='IE')
    plt.plot(tMM,yMM,'r-',label='MM')
    plt.xlabel('time')
    plt.ylabel('y(t)')
    leg=plt.legend(); leg.draw_frame(False)
    plt.title('dt=' + str(dt[idt]))
    plt.gca().set_ylim([1,4])
    plt.gca().set_xlim([0,8])
plt.tight_layout()


```

Note, MM is much better than EE, IE.  EE and IE split the exact solution, and the approximation improves as the stepsize decreases.  The MM is pretty close (visually) for all stepsizes.

# Problem 3

*Solve the following ODE using the explicit Euler method:
$$\begin{align}&\\ \frac{dy}{dt} &= -y, \\
y(t=0) &= 1.\end{align}$$
Solve the equation from $t = 1$ to $t = 20$ using 9, 10, 15, 19, 20, 21, 30, and 60 steps. Make 8 plots (a
4x2 subplot) with the result and exact solution on each plot for each of the 8 steps. That is, for example,
subplot(1,1) has the solution ploted with the exact solution for 9 steps (2 curves on the plot), and you have
8 plots like this. Oscillations are normal for some of these plots!*

*Here is a template you can use if you want.* (Solution added to the original template.)

```python
def EEstep(t,dt,yn,rhsf) :
    return yn + dt * rhsf(t,yn)

def p3_rhsf(t,y) :
    return -y
```

```python
#---------------- Setup the problem

T = 20                  # end time
y0 = 1.0                # initial condition
Ns = np.array([9,10,15,19,20,21,30,60])  # number of timesteps to do for each of 8 runs
te = np.linspace(0,T,1000)  # domain for plotting the exact solution
ye = y0*np.exp(-te)         # enter the exact solution

plt.subplots(figsize=(10,10))
plt.tight_layout()

#--------------- solve the problem

for iNs in range(8) :

    dt = T/Ns[iNs]          # step size
    Np = Ns[iNs] + 1        # number of points from t=0 to T
    t = np.linspace(0,T,Np) # time array
    y = np.zeros(Np)        # solution array
    y[0] = y0               # fill in the initial condition

    for n in range(Ns[iNs]) :
        y[n+1] = EEstep(t[n],dt,y[n],p3_rhsf)

    #--------------- plot the results

    plt.subplot(4,2,iNs+1)
    plt.plot(te,ye,'--',label='exact')
    plt.plot(t,y,'-', label='EE')
    plt.title('dt = '+str(dt))
    if iNs >= 6 : 
        plt.xlabel('t')
    if iNs == 7 :
        leg = plt.legend(); 
        leg.draw_frame(False)
    
    plt.ylabel('y(t)')
    if iNs == 0 :
        leg = plt.legend()
        leg.draw_frame(False)
```

The plots are shown above: From left to right and top to bottom the step size decreases. In the first plot, the solution oscillates and the error grows. In the second plot, the solution oscillates, but the error does not grow. In the third and fourth plots, the solution oscillates but the solution is a stable approximation of the true result. In the fifth plot, there is no oscillation. This timestep size is the threshold above which
oscillation occurs. Plots 6, 7, and 8 show no oscillation and an improved solution as the stepsize decreases. Plot 2  has a timestep that is stable, but the step size is the threshold above which the solution explodes.

# Problem 4


*Solve the ODE of problem 1 using the RK4 method to a time of t = 8. Vary the step size and verify the order of the method by comparing the error at t = 8 for various step sizes. This is best done graphically. On log scales you’ll get a slope indicative of the order of the method.*

```python
def odeRK4(y0, frhs, Tend, dt) :
    """ 
    Solve an ode problem dy/dt = frhs(y,t) from t=0 to t=Tend using stepsize dt.
    y0 is the initial condition.
    Uses the RK4 method 
    returns y(t),t
    """
    
    nsteps = int(Tend/dt)
    ntimes = nsteps + 1
    t = np.linspace(0.0,Tend,ntimes)
    y = np.zeros(ntimes)
    y[0] = y0
        
    for n in range(nsteps) :
        
        dy1 = dt*frhs(y[n],         t[n])
        dy2 = dt*frhs(y[n]+dy1*0.5, t[n]+0.5*dt)
        dy3 = dt*frhs(y[n]+dy2*0.5, t[n]+0.5*dt)
        dy4 = dt*frhs(y[n]+dy3,     t[n]+dt)
        
        y[n+1] = y[n] + 1.0/6.0*( dy1 + 2.0*dy2 + 2.0*dy3 + dy4)
    
    return y,t
```

```python
def p4_rhsf(y,t) :
    return y*np.exp(-t)
```

## Now run the code

```python
#------------- setup

nrun = 10                  # number of step sizes to do
y0 = 1.0                   # initial condition
tend = 8.0                 # integration end time
Err = np.zeros(nrun)       # initialize the error array at t = 8.0
ii = np.arange(1,nrun+1)   # index array
NS = 2.0**ii               # array of the number of steps
DT = tend/NS               # step size array
ye = y0*np.exp(1.0-np.exp(-tend)) # exact solution at tend

#------------- Solve for each run (each timestep size)

for i in range(nrun) :
    y,t = odeRK4(y0, p4_rhsf, tend, DT[i])      # integrate the ODE to find y(t=tend)
    Err[i] = np.abs(y[-1] - ye)              # store the error for plotting
    
#-------------- plot results

plt.plot(DT,Err,'bo-')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Step size, dt')
plt.ylabel('Abs Error at t=8')
plt.gca().set_ylim([1E-12,1])

#-------------- Slope

p = np.polyfit(np.log(DT[-3:]),np.log(Err[-3:]),1)
print('Slope of the line through the last three points = ', p[0])
```

Note, the slope of log(Error) versus log(dt) = 4, which is consistent with the 4th order method.
In general, an $O(\Delta t^k)$ method, has $\epsilon = \alpha \Delta t^k$, where $\alpha$ is some constant.  Then a plot of $\log(\epsilon)$ versus $\log(\Delta t^k)$ will be linear with slope $k$.
