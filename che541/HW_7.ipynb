# Problem 2

*The temperature profile of a nonpremixed flame can be approximated by
$$\frac{\partial T}{\partial t} = \frac{\chi}{2}\frac{\partial^2T}{\partial\xi^2} + S(\xi).$$

Here, $S$ is a reactive source term. The left boundary $\xi=0$ corresponds to pure air, and the right boundary $\xi=1$ corresponds to pure fuel. The source term is really also a function of composition requiring the solution of several more ODE’s simultaneously. The source term given is derived from that solution. Solve this equation on the domain $0\le\xi\le 1$ with T = 298.15 K at both boundaries. Solve the problem from t = 0 to t = 0.001. A driver program is provided, that also gives the initial temperature profile to use. Change parts of the program (or everything if you like) that are marked in upper case comments. You will have to add external functions as needed. (Note, in the code below $\xi\equiv f$.)

## Part a

Solve the problem using the explicit FTCS method. Plot the solution as a function of time. That is, plot $T$ versus $\xi$ for various times on a single plot. Use a stepsize of 8.3333E-8 (1200 timesteps).


```python
dt = 8.3333e-8
dx = xi[1]-xi[0]
Chi = chia
def fNext(f,t,Chi,S,xi):
    N = np.size(f)
    fnew = f.copy()
    i = np.arange(1,N-1)
    fnew[i] = f[i] + Chi[i]/2 * (f[i-1] - 2*f[i] + f[i+1])/dx**2*dt + S(xi[i])*dt
    return fnew
fValues = np.array([T0])
time = np.array([0])
tEnd = 0.001
while time[-1]+dt < tEnd:
    fValues = np.vstack((fValues,fNext(fValues[-1,:],time[-1],Chi,S,xi)))
    time    = np.vstack((time,time[-1]+dt))
```

```python
def strf(i):
    return 't={:.1g}'.format(i[0])
lblstring=list(map(strf, time[::2400]))
#plt.plot(fValues[:,::1200].T)
muffle=plt.plot(fValues[::2400,:].T)
plt.legend(lblstring,loc='upper right')
plt.xlabel('time (s)')
muffle=plt.ylabel('T (Kelvin)')
```

## Part b

Solve the problem using the implicit BTCS method with 10 timesteps). Plot the solution as a function of time. That is, plot T versus $\xi$ for various times on a single plot.


```python
timeb = np.linspace(0,0.001,10,endpoint=True)
dt = timeb[1]-timeb[0]
def BTCS(f,t,Chi,S,xi):
    N = np.size(f)
    fnew = f.copy()
    i = np.arange(1,N-1)
    def nonlinearFunc(fnew):
        fn = fnew.copy()
        fn[i] = -(fnew[i] - f[i])/dt + Chi[i]/2 * (fnew[i-1] - 2*fnew[i] + fnew[i+1])/dx**2 + S(xi[i])
        return fn
    fn = sco.fsolve(nonlinearFunc,fnew)
    return fn
fValuesb = np.array([T0])
tEnd = 0.001
for i in timeb:
    if i!= 0:
        fValuesb = np.vstack((fValuesb,BTCS(fValuesb[-1,:],i,Chi,S,xi)))
```

```python
def strf(i):
    return 't={:.1g}'.format(i)
lblstring=list(map(strf, timeb))
plt.plot(fValuesb.T)
plt.legend(lblstring,loc='best')
plt.xlabel('time (s)')
muffle=plt.ylabel('T (Kelvin)')
```

## Part c

Solve the problem using the Crank Nicholson method with 10 timesteps). Plot the solution as a function of time. That is, plot T versus $\xi$ for various times on a single plot. Report your stepsize.


```python
timec = np.linspace(0,0.001,10,endpoint=True)
dt = timec[1]-timec[0]
def CNrhs(f,t,Chi,S,xi,omega=0.5):
    N = np.size(f)
    fnew = f.copy()
    i = np.arange(1,N-1)
    lower = Chi[:-1]*omega/(2*dx*dx)
    upper = Chi[1:]*omega/(2*dx*dx)
    middle = (1/dt+Chi[:]*omega/(2*dx*dx))
    b = f[i]/dt+(1-omega)*(f[i+1]-2*f[i]+f[i-1])*Chi[i]/(2*dx*dx)+S(xi[i])
    fn = thomas(lower,middle,upper,b)
    return fn

fValuesc = np.array([T0])
tEnd = 0.001
for i in timec:
    if i!= 0:
        fValuesc = np.vstack((fValuesc,CNrhs(fValuesc[-1,:],i,Chi,S,xi)))
```
