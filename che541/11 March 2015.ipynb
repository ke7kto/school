PDE's: 

1. Parabolic
    - $\d{f}{t} = \alpha\dtwo{f}{x} + \alpha\dtwo{f}{y} + S$
    - 1-D 
    - 2-D 
    - ADI
    - FTCS
    - BTCS
    - Omega method

2. Elliptic
    - $0 = \alpha\dtwo{f}{x} + \alpha\dtwo{f}{y} + S$ 
    - Finite volume solutions
    - No time involvement
    - Diffusion-like
    - "No" Marching
    - "1 solution"
    - $\frac{\alpha}{\Delta x^2} \left(f_{i-1,j} - 2f_{i,j} + f_{i+1,j}\right) + \frac{\alpha}{\Delta y^2}\left(f_{i,j-1} - 2f_{i,j}+f_{i,j+1}\right) + S_{i,j}= 0$
    - Common approach is to solve the unsteady problem to a steady state solution.
    
3. Hyperbolic

- Gauss Seidel knocks out high frequency noise quickly, but struggles to eliminate low frequency noise.
- Can't see the forest for the trees? Use multi-grid methods! (The change between step sizes makes low frequency noise look like high frequency noise)

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from matplotlib import animation

N = 41
nx = N
ny = N
h = 0.01
f = np.random.random((nx,ny))-0.5
x = np.linspace(0,1,nx)
y = np.linspace(0,1,ny)
X,Y = np.meshgrid(x,y)
def F(i,j):
    return i+j-i*i
fig=plt.figure()
ax = plt.gca(projection='3d')
print(f)
line=ax.plot_surface(X,Y,f)
print(line)
```
