$$\begin{align}&\\ \frac{dy}{dt} &= ye^{-t}, \\
y(t=0) &= 1.\end{align}$$*

```python
def f(y,t):
    return y*exp(-t)
def fExact(t):
    #ln(y)= 1-e^-t
    return np.exp(1-np.exp(-t))
fE = np.vectorize(fExact)
def EE(fPrime,y0,tstep,tend):
    t = [0]
    yList =[y0]
    while t[-1] < tend:
        yList.append(yList[-1]+tstep*fPrime(yList[-1],t[-1]))
        t.append(t[-1]+tstep)
    return yList,t
def IE(fDummy,y0,tstep,tend):
    t = [0]
    yList =[y0]
    while t[-1] < tend:
        #yNew = yOld + tstep*yNew*exp(-(t+tstep))
        #yNew(1 - tstep*exp(-t+tstep)) = yOld
        #yNew = yOld/(1-tstep*exp(-(t+tstep)))
        yList.append(\
             yList[-1]/(1-tstep*np.exp(-(t[-1]+tstep))\
        ))
        t.append(t[-1]+tstep)
    return yList,t
def MMidpoint(fPrime,y0,tstep,tend):
    t = [0]
    yList =[y0]
    while t[-1] < tend:
        yTest = yList[-1]+tstep/2*fPrime(yList[-1],t[-1])
        yList.append(yList[-1]+tstep*fPrime(yTest,t[-1]+tstep/2))
        t.append(t[-1]+tstep)
    return yList,t
#def 
y,t = EE(f,1, .2,8)
y2,t2=EE(f,1, .4,8)
y3,t3=EE(f,1, .8,8)
y4,t4=EE(f,1,1.6,8)
yi,ti=  IE(f,1, .2,8)
yi2,ti2=IE(f,1, .4,8)
yi3,ti3=IE(f,1, .8,8)
yi4,ti4=IE(f,1,1.6,8)
ym,tm=  MMidpoint(f,1, .2,8)
ym2,tm2=MMidpoint(f,1, .4,8)
ym3,tm3=MMidpoint(f,1, .8,8)
ym4,tm4=MMidpoint(f,1,1.6,8)
tE = np.arange(0,8,.2)
tE2= np.arange(0,8,.4)
tE3= np.arange(0,8,.8)
tE4= np.arange(0,8,1.6)
plt.subplot(2,2,1)
plt.title('$\\Delta$ t = 0.2')
plt.plot(t , y,label='EE')
plt.plot(ti,yi,label='IE')
plt.plot(tm,ym,label='MM')
plt.plot(tE,fE(tE),label='Exact')
plt.legend(loc='upper center',bbox_to_anchor=(1.1,1.550),ncol=2)
plt.subplot(2,2,2)
plt.title('$\\Delta$ t = 0.4')
plt.plot(t2,y2,label='EE')
plt.plot(ti2,yi2,label='IE')
plt.plot(tm2,ym2,label='MM')
plt.plot(tE,fE(tE),label='Exact')
plt.subplot(2,2,3)
plt.title('$\\Delta$ t = 0.8')
plt.plot(t3,y3,label='EE')
plt.plot(ti3,yi3,label='IE')
plt.plot(tm3,ym3,label='MM')
plt.plot(tE,fE(tE),label='Exact')

plt.subplot(2,2,4)
plt.title('$\\Delta$ t = 1.6')
plt.plot(t4,y4,label='EE')
plt.plot(ti4,yi4,label='IE')
plt.plot(tm4,ym4,label='MM')
plt.plot(tE,fE(tE),label='Exact')
plt.tight_layout()
```

# Problem 3

*Solve the following ODE using the explicit Euler method:
$$\begin{align}&\\ \frac{dy}{dt} &= -y, \\
y(t=0) &= 1.\end{align}$$
Solve the equation from $t = 1$ to $t = 20$ using 9, 10, 15, 19, 20, 21, 30, and 60 steps. Make 8 plots (a
4x2 subplot) with the result and exact solution on each plot for each of the 8 steps. That is, for example,
subplot(1,1) has the solution ploted with the exact solution for 9 steps (2 curves on the plot), and you have
8 plots like this. Oscillations are normal for some of these plots!*

*Here is a template you can use if you want.* (Solution added to the original template.)

```python
#Note for Chen:
#I decided that the question asked me to solve the equation from t=1 to t=20, but since I was given y(t=0)
#I just did one leap from t=0 to t=1. I thought that was fair to all of the steps, but it means that every function
#will be off at the beginning.

def df(y,t):
    return -y
def ExplicitEuler3(df,y0,tstart,tend,nSteps):
    y = []
    t = [tstart]
    y.append(df(y0,0)*tstart)
    dt = (tend - tstart)/nSteps
    while t[-1]<tend:
        y.append(y[-1] + df(y[-1],t[-1])*dt)
        t.append(t[-1]+dt)
    return y,t
def exact(t):
    return -np.exp(-t)
exact = np.vectorize(exact)
steps = [9,10,15,19,20,21,30,60]
plt.figure(figsize=(10,5),dpi=300)
for i in range(len(steps)):
    plt.subplot(4,2,i+1)
    y,t = ExplicitEuler3(df,1,1,20,steps[i])
    plt.plot(t,y,'b-',label='Appx')
    plt.plot(t,exact(t),'g--',label='Exact')
    plt.title('{:d} steps'.format(steps[i]))
    if i > 2:
        plt.ylim([-1,0.4])
plt.subplot(4,2,1)
plt.legend(loc='upper center',bbox_to_anchor=(1.125,2),ncol=2)

plt.tight_layout()
```

```python
def RungaKutta(fPrime,y0,tstep,tend):
    y = [y0]
    t = [0]
    while t[-1]<tend:
        dy1 = tstep*fPrime(y[-1],t[-1])
        dy2 = tstep*fPrime(y[-1]+dy1/2,t[-1]+tstep/2)
        dy3 = tstep*fPrime(y[-1]+dy2/2,t[-1]+tstep/2)
        dy4 = tstep*fPrime(y[-1]+dy3  ,t[-1]+tstep)
        t.append(t[-1]+tstep)
        y.append(\
                 y[-1]+1/6.*(dy1+2*dy2+2*dy3+dy4)\
                 )
    return y,t
tstep = [8,4,2.,1.,0.5,0.25,0.125,.0625]
for i in tstep:
    y,t=RungaKutta(f,1,i,8)
    plt.semilogy(i,np.abs(y[-1]-fE(t[-1]))/fE(t[-1]),'bo')
```
