Quadrature

$I = \Sigma f_i \cdot w_i$<br/>
Gauss Quadrature:<br/>
$\eta_i \:|\: w_i \:|\: f(n_i) \rightarrow \Sigma$

```python
from __future__ import division
import matplotlib.lines as mpl
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / 2 * np.power(sig, 2.))
x = np.linspace(0,1,100)
plt.plot(x,gaussian(x,0.5,10))
count=0
for i in x:
    if count%5==0:
        plt.plot([i,i],[0,gaussian(i,0.5,10)])
    count+=1
```

1. Trapezoid Rule
2. $I_1 = \frac{b-a}{2}\cdot(f(b)-f(a))$
3. Midpoint rule:
4. $c=\frac{b+a}{2}$
4. $I_2 = \frac{c-a}{2}\cdot(f(c)-f(a))+\frac{b-c}{2}(f(b)-f(c))$
6. Compare $I_1$ to $I_2$
    - Same: Don't refine further
    - Different: Repeat step 4.

```python
TOL = 1e-6
RTOL = 1e-6
def quadMe(f,a,b,ax=None):
    I = (b-a)/2*(f(b)+f(a))
    c = (b+a)/2
    I2 = (c-a)/2*(f(c)+f(a))
    I3 = (b-c)/2*(f(b)+f(c))    
    if (abs(I-(I2+I3))<TOL):
        return I
    else:
        lower =quadMe(f,a,(a+c)/2,ax)+quadMe(f,(a+c)/2,c,ax)
        upper = quadMe(f,c,(b+c)/2,ax)+quadMe(f,(b+c)/2,b,ax)
        if type(ax) is not type(None):
            l2 = mpl.Line2D([b,b],[0,upper+lower])
            ax.add_line(l2)
        return lower+upper
quadMe = np.vectorize(quadMe)
```

```python
def f(x):
    return x**2
```

```python
x = np.linspace(0,1,100)
fig=plt.figure(figsize=(12,5))
fig.add_subplot(1,2,1)
ax = fig.gca()
plt.plot(x,1/3*x**3,label='exact')
quadMe(f,0,1,ax)
plt.legend(loc='best')
fig.add_subplot(1,2,2)
plt.plot(x,quadMe(f,0,x)-1/3*x**3,label='error')
plt.legend()
```
