<h1>Parabolic PDE's</h1>
PDE types:
$\newcommand{\dd}[2]{\frac{\partial #1}{\partial#2}}$
$\newcommand{\ddd}[3]{\frac{\partial^#1 #2}{\partial#3^#1}}$

1. Parabolic
  - Space dependence is everywhere
  - Time dependence is only dependent on the history
  - $\dd{T}{t} + \ddd{2}{T}{x} = 0$
  - $\dd{f}{t} + u\dd{f}{x} + D\ddd{2}{f}{x} = S$ is a parabolic equation.
  - Ex: Navier-Stokes
  - Next-closest thing to ODE-land.
2. Hyperbolic
  - Convection
  - characteristic curves exist where the value of the function is always the same as it was before, but shifted in distance as we march through time
  - Depend locally on space
3. Elliptic
  - Diffusive (felt everywhere
  - Steady state (no time dependence)
  - $\dd{^2T}{x^2}=0$ is elliptic


- $\ddd{2}{f}{x}=$Use central difference
- $\dd{f}{t} + \dd{(uf)}{x}=\dd{f}{t} + u\dd{f}{x} =$a clear direction of propagation. Central difference will blow up miserably. Use a one-sided (upwind) difference Ex $\frac{f_{i+1} - f_{i}}{\Delta x}$
- $\dd{f}{t} + u\dd{f}{x} + D\ddd{2}{f}{x} = S$ is a parabolic equation.
- Take terms as follows:$$\dd{f}{t} + u\dd{f}{x}\text{(1-sided [convective])} + D\ddd{2}{f}{x}\text{(Central [diffusive])} = S$$


<h3>Unsteady Heat equation</h3>
1. $\dd{f}{t}  = \alpha\ddd{2}{f}{x}$
2. Discretize space:
 - in some domain L, for some point $i$: $\dd{f}{t} = \frac{\alpha}{\Delta x^2} \left(f_{i-1}-2f_i + f_{i+1}\right)$
 - This is called the method of lines
 - Feed this into an ode solver for each $i$ and you're done!


Or explicit Euler
$f_i^{n+1} = f_i^n + \frac{\Delta t \alpha}{\Delta x^2} \left(f_{i-1}^n - 2 f_i^n + f_{i+1}^n\right) - \frac{\Delta t\alpha}{\Delta x^2} (25\Delta x)$
This is the Forward-time centered space algorithm

There is an analog to FTCS with implicit euler:
$f_i^{n+1} = f_i^n + \frac{\Delta t \alpha}{\Delta x^2} \left(f_{i-1}^{n+1} - 2 f_i^{n+1} + f_{i+1}^{n+1}\right) - \frac{\Delta t\alpha}{\Delta x^2} (25\Delta x)$


Crank-Nicholson:
$f_{n+1} = f_n + \frac{\Delta t}{2}\cdot \left(RHS^{n+1} + RHS^{n}\right)$

Omega method:
$f_{n+1} = f_n + \frac{\Delta t}{2}\cdot \left(\omega RHS^{n+1} +(1-\omega) RHS^{n}\right)$
