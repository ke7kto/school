*The 1-D convection equation is written as:
$$\frac{\partial u}{\partial t} + c\frac{\partial u}{\partial x},$$
where $c$ is the wave speed.  We will solve this PDE on the x-domain from x=0 to x=13 with periodic boundary conditions.  The problem will be solved for one "flow through" time.*  


*You should use the following initial condition in the code below.  This gives four peaked reagions with domains (of x): [0,4], [4,6], [7,9], [10,12].  
The functions on these regions are* 
* $u=\sech(8x-2)$, 
* $u = 1-(x-5)^2$, 
* a triangle with a peak of $u=1$ at $x=8$ and a base with $u=0$ at $x=7$ and $x=9$, and 
* a tophat with $u=1$, respectively.

Solve the problem using the  <strike>upwind,second order upwind, and Lax Wendroff methods</strike>. Compare the results graphically at the final time. Include the exact solution at the final time (which is just the initial condition). I recommend just using a single code and stepping three different versions of the solution (in three different arrays), then plot your three methods on one graph.

You should vary the CFL and the grid size. You can verify your code by checking against a CFL of 1. Use CFL=0.3, 0.5, 0.7, and 0.9 with 100 grid points, and use CFL=0.9 with N=50, 100, 300, and 500 grid points. You will have 8 plots.

Comment on which method you would use and why.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym

Ld  = 13.0
nx  = 100
c   = 1.0
cfl = 0.9
ngrid = nx
cfls = [0.3,0.5,0.7,0.9,1]
Ngrids=[50,100,300,500,1000]
dx = Ld/nx     # not Ld/(nx-1)
x = np.linspace(0.0, Ld-dx, nx)

u0 = np.zeros(nx)
SOunp1 = u0.copy()
FOunp1 = u0.copy()
LAXnp1 = u0.copy()

i = np.where((x>=0) & (x<=4));   u0[i] = 1.0/(np.cosh(8.0*(x[i]-2.0)))
i = np.where((x>=4) & (x<=6));   u0[i] = 1.0-(x[i]-5.0)**2.0
i = np.where((x>=7) & (x<=8));   u0[i] = x[i]-7.0
i = np.where((x>=8) & (x<=9));   u0[i] = 1.0-(x[i]-8.0)
i = np.where((x>=10)& (x<=12));  u0[i] = 1.0
fig=plt.figure(figsize=(18,19))
Nplot=1
for cfl in cfls:
    #Make the plots for the different cfl's
    #set up the proper timesteps based on cfl
    dt = cfl*dx/c
    tend = Ld/c
    tsteps = np.arange(0,tend,dt)
    SOunp1 = u0.copy()
    FOunp1 = u0.copy()
    LAXnp1 = u0.copy()
    i          = np.arange(ngrid)
    iminus2     = i -2
    iminus2[0] = i[-2]
    iminus2[1] = i[-1]# wrap around the periodic
    iminus1    = i - 1
    iminus1[0] = i[-1]           # wrap around the periodic 
    iplus1 = i + 1
    iplus1[-1]=i[0]
    for t in tsteps:
        SOu = SOunp1.copy()
        #This is the second order upwind
        SOunp1[i] = SOu[i] - cfl*(SOu[i]-SOu[iminus1])-cfl*(1-cfl)/2*(SOu[i]-2*SOu[iminus1]+SOu[iminus2])
        FOu = FOunp1.copy()
        #First order upwind
        FOunp1[i] = FOu[i]- cfl*(FOu[i]-FOu[iminus1])
        LAX = LAXnp1.copy()
        #Lax-Wendroff one-step method
        LAXnp1[i] = LAX[i]-cfl/2*(LAX[iplus1]-LAX[iminus1])+cfl*cfl/2*(LAX[iplus1]-2*LAX[i]+LAX[iminus1])
    fig.add_subplot(2,5,Nplot)
    Nplot+=1
    #Plot it!
    plt.plot(x,u0,'.-',label='Exact Solution')
    plt.plot(x,FOu,'.-',label='UW 1st Order')
    plt.plot(x,SOu,'.-',label='UW 2nd Order')
    plt.plot(x,LAX,'.-',label='LAX-Wendroff')
    plt.gca().set_xlim([0,13])
    plt.gca().set_ylim([-0.2,1.75])
    
    plt.title('CFL={:.1f}, Ngrid={:.1f}'.format(cfl,ngrid))
    plt.xlabel('t')
    plt.ylabel(r'u(t)')
    plt.legend(loc='best')
cfl = 0.9
#Make the plots with different spatial grids
for ngrid in Ngrids:
    #Make the appropriate stepsize
    dx = Ld/ngrid     # not Ld/(nx-1)
    x = np.linspace(0.0, Ld-dx, ngrid)
    #Make u0 [u(0)] the right size 
    u0 = np.zeros(ngrid)
    SOunp1 = u0.copy()
    FOunp1 = u0.copy()
    LAXnp1 = u0.copy()

    i = np.where((x>=0) & (x<=4));   u0[i] = 1.0/(np.cosh(8.0*(x[i]-2.0)))
    i = np.where((x>=4) & (x<=6));   u0[i] = 1.0-(x[i]-5.0)**2.0
    i = np.where((x>=7) & (x<=8));   u0[i] = x[i]-7.0
    i = np.where((x>=8) & (x<=9));   u0[i] = 1.0-(x[i]-8.0)
    i = np.where((x>=10)& (x<=12));  u0[i] = 1.0
    #Define dt by cfl, otherwise overspecified
    dt = cfl*dx/c
    tend = Ld/c
    tsteps = np.arange(0,tend,dt)
    SOunp1 = u0.copy()
    FOunp1 = u0.copy()
    LAXnp1 = u0.copy()
    i          = np.arange(ngrid)
    iminus2     = i -2
    iminus2[0] = i[-2]
    iminus2[1] = i[-1]# wrap around the periodic
    iminus1    = i - 1
    iminus1[0] = i[-1]           # wrap around the periodic 
    iplus1 = i + 1
    iplus1[-1]=i[0]
    #March it!
    for t in tsteps:
        
        SOu = SOunp1.copy()
        #Second-order upwinding
        SOunp1[i] = SOu[i] - cfl*(SOu[i]-SOu[iminus1])-cfl*(1-cfl)/2*(SOu[i]-2*SOu[iminus1]+SOu[iminus2])
        FOu = FOunp1.copy()
        #First-order upwinding
        FOunp1[i] = FOu[i]- cfl*(FOu[i]-FOu[iminus1])
        LAX = LAXnp1.copy()
        #Lax-Wendroff
        LAXnp1[i] = LAX[i]-cfl/2*(LAX[iplus1]-LAX[iminus1])+cfl*cfl/2*(LAX[iplus1]-2*LAX[i]+LAX[iminus1])
    fig.add_subplot(2,5,Nplot)
    Nplot+=1
    #Plot it!
    plt.plot(x,u0,'.-',label='Exact Solution')
    plt.plot(x,FOu,'.-',label='UW 1st Order')
    plt.plot(x,SOu,'.-',label='UW 2nd Order')
    plt.plot(x,LAX,'.-',label='LAX-Wendroff')
    plt.gca().set_xlim([0,13])
    plt.gca().set_ylim([-0.2,1.65])
    
    plt.title('CFL={:.1f}, Ngrid={:.1f}'.format(cfl,ngrid))
    plt.xlabel('t')
    plt.ylabel(r'u(t)')
    plt.legend(loc='best')
```
