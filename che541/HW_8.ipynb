*In 2-D we have $f_t = \alpha f_{xx} + \alpha f_{yy}$.  If we assume $\Delta x = \Delta y$, the FTCS equation is given by
$$ f_{i,j}^{n+1} = f_{i,j}^n + d(f_{i,j-1}^n +f_{i-1,j}^n-4f_{i,j}^n+f_{i+1,j}^n+f_{i,j+1}^n).$$*

For the stability analysis, we use $f = \sum_m c_me^{-ik_mx}$.  In 2-D we have the following:
$$f_{i,j}^n = \sum_mc_me^{-i\vec{k_m}\cdot\vec{x_{i,j}}} 
=\sum_mc_me^{-i(k_{m,x}x_i+k_{m,y}y_j)}
=\sum_mc_me^{-ik_{m,x}x_i}e^{-ik_{m,y}y_j}.
$$

The PDE and its descrete FDCS equation are both linear, so we can analyse the stability in terms of each term in the sum separately.  Note, that a term like $f_{i+1,j}$ is evaluated at $x_{i}+\Delta x$, so, for that term (and for a single "mode" in the sum) we have
$$f_{i+1,j}^n = e^{-ik_{m,x}x_{i+1}}e^{-ik_{m,y}y_j}
= e^{-ik_{m,x}x_i+\Delta x}e^{-ik_{m,y}y_j}
= e^{-ik_{m,x}x_i}e^{-ik_{m,y}y_j}e^{-ik_{m,x}\Delta x}
= f_{i,j}^ne^{-ik_{m,x}\Delta x}
$$
Again, the first and last equalities give:
$$f_{i+1,j}^n =f_{i,j}^ne^{-ik_{m,x}\Delta x}.$$

So, each term in the discrete equation is just $f_{i,j}$ times $e^{\pm k_{m,x\_or\_y}\Delta x\_or\_\Delta y}$.  That is the "mode" is an eigenfunction of the PDE and its discrete representation. We make the substitution, Then divide through by $f_{i,j}^n$ to get:

$$
\frac{f_{i,j}^{n+1}}{f_{i,j}^{n+1}} = 1+d\cdot[
e^{ik_{m,y}\Delta y} +
e^{ik_{m,x}\Delta x} +
-4
e^{-ik_{m,x}\Delta x} +
e^{-ik_{m,y}\Delta y}
].
$$

Now use the Euler identity: $e^{ix} = \cos(x) + i\sin(x)$, to get

$$
\frac{f_{i,j}^{n+1}}{f_{i,j}^{n+1}} = 1 + d\cdot[2\cos(k_{m,y}\Delta y) + 2\cos(k_{m,x}\Delta x) - 4].
$$

For stability, we want the magnitutde of this quantity to be $\le 1$. Also, $-1\le\cos(\theta)\le 1$.  We only need to consider the bounds, so we can replace the $\cos$ in the above equation with its bounds:

$$
-1 \le 1+d\cdot[\pm 2 \pm 2 - 4] \le 1.
$$

The upper bound is always satisfied since the largest the the term in brackets can be is 0, which is always $\le 1$.  For the lower bound
we have three cases for the term in brackets [+2 +2 -4]=0, [+2 -2 -4]=-4 (or [-2 +2 -4]=-4), and [-2-2-4]=-8, and the general stability criteria is the most stringent case.  
1. $-2\le d*0$ holds for any $d$
2. $-2 \le d*-2$ gives $d\le 1/2$
3. $-2\le d*-4$ gives $d\le 1/4$.

**So, 
$d\le 1/4$
is the 2-D stability criteria.**

# Problem 2

Two dimensional laminar fluid flow may be conveniently described using a vorticity-stream function formulation. The vorticity equation is derived from the momentum equation by taking the curl $\nabla\times$ of the momentum equation. The result in 2D is
$$ 
\begin{equation}
\frac{\partial \omega}{\partial t} = 
-u\frac{\partial \omega}{\partial x} 
-v\frac{\partial \omega}{\partial y}
+ \nu\frac{\partial^2 \omega}{\partial x^2}
+ \nu\frac{\partial^2 \omega}{\partial y^2},
\end{equation}
$$
where 
$$
\omega = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y}.
$$
Note, $u$ is the x velocity component, $v$ is the y velocity component and $\nu=\mu/\rho$ is kinematic viscosity.

The stream function is parallel to velocity streamlines and is defined as 
$$
\begin{align}
\frac{\partial\psi}{\partial y} &= u, \\
\frac{\partial\psi}{\partial x} &= -v.
\end{align}
$$
The stream function is related to the vorticity by 
$$
\frac{\partial^2 \psi}{\partial x^2}+
\frac{\partial^2 \psi}{\partial y^2}
= -\omega.
$$

Hence, 2-D fluid flow can by computed by solving the first and last equations.

A very common test problem is the lid driven cavity consisting of a square chamber will wall boundaries. The lid of the chamber moves to the right with a velocity $U_{bc}$. Otherwise all velocity components on the walls are zero. The square boundaries are themselves one of the streamlines, so that the stream function is constant on the boundaries. From the above equations it is clear that the stream function is defined only up to a constant, so the stream function may be conveniently defined to be zero on the boundaries. For the vorticity boundary conditions, a Taylor series evaluated at the wall using a base point one grid cell away from the wall is applied to the stream function including the first three terms of the series. This gives an equation for the wall vorticity in terms of the wall velocity, and the wall stream function, and the stream function one point from the wall.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time
from __future__ import division
from IPython import display
import scipy.sparse
import scipy.sparse.linalg
```

```python
class fluid_solver :
    
    def __init__(self):
        
        self.plotmod=200      # # steps between plot output
        
        Lx = 1.               # domain size: x
        Ly = 3.               # domain size: y
        Re = 100.             # Reynolds number: use with velocity, L to set nu
        
        self.nx = 21          # # grid points including boundaries
        self.ny = 21
        
        self.uBCT = 1.0       # Top velocity
        self.uBCB = 0.0       # Bottom velocity
        self.vBCW = 0.0       # West
        self.vBCE = 0.0       # East
 
        self.nsteps  = 1000   # # of timesteps
        self.cfl     = 0.25   # timestep size factor 

        self.nu = np.max(np.abs([self.uBCT,self.uBCB,self.vBCW,self.vBCE]))*Lx/Re # mu/rho

        #--------- set the grid
    
        x = np.linspace(0,Lx,self.nx)     
        y = np.linspace(0,Ly,self.ny)

        self.dx = x[1] - x[0]
        self.dy = y[1] - y[0]

        self.X,self.Y = np.meshgrid(x,y)    # for plotting

        self.dt = self.cfl * np.min([self.dx,self.dy])/np.max(np.abs([self.uBCT,self.uBCB,self.vBCE,self.vBCW]))
    
        self.times = np.linspace(0.0,self.nsteps*self.dt,self.nsteps+1)
    
        #---------- set u, v, omega, psi fields
    
        self.u = np.zeros((self.nx,self.ny))
        self.v = np.zeros((self.nx,self.ny))

        self.u[:,0]  = self.uBCB   # velocity boundary conditions
        self.u[:,-1] = self.uBCT
        self.v[0,:]  = self.vBCW
        self.v[-1,:] = self.vBCE

        self.w = np.zeros((self.nx,self.ny))    # vorticity (omega)
        self.P = np.zeros((self.nx,self.ny))    # stream function (P as in psi)
        
        return
    
    #=========================================================
        
    def driver(self) :
                
        self.setWfromV()                           # init omega from init u, v fields
    
        self.Ptrc  = np.zeros(self.nsteps+1)       # keep track of norm(psi) to watch convergence
    
        for istep in range(1,self.nsteps+1) :      # loop over timesteps
        
            self.Ptrc[istep] = np.linalg.norm(self.P)   # grab the Psi trace
        
            self.w = self.w + self.dt*self.rhsf()  # step the solution
            
            #-------- plot results

            if istep%self.plotmod==0 :
                self.plotCurrentSolution()
                
        display.clear_output(wait=True)             # so we don't get two plots when we're done
        
        return
        
    #======================================================
    
    def plotCurrentSolution(self) :
        
        plt.clf()
                
        plt.subplot(2,2,1)
        plt.contour(self.X,self.Y,self.P.T,60) 
        plt.title('Stream Function')
                
        plt.subplot(2,2,2)
        plt.contour(self.X,self.Y,self.w.T,60) 
        plt.title('Vorticity')
   
        self.setUVfromP()
        plt.subplot(2,2,3)
        plt.contourf(self.X,self.Y,self.u.T,60)
        plt.title('u velocity');
    
        plt.subplot(2,2,4)
        plt.plot(self.times[0::10], self.Ptrc[0::10], 'o-')
        plt.xlabel('Time')
        plt.ylabel(r'$||\psi||_2$')
        
        plt.tight_layout()
    
        display.clear_output(wait=True)
        display.display(plt.gcf())
        #time.sleep(0.1)

    #======================================================

    def setWfromV(self) :

        """Get the vorticity field w from the velocity field u, v."""
    
        u = self.u
        v = self.v
        w = self.w
        dx = self.dx
        dy = self.dy
        nx = self.nx
        ny = self.ny
        ix_ = np.ix_         # for convenience, so we can write ix_ instead of np.ix_
        
        #--------------- Interior points
       
        i = np.arange(1,nx-1); j = np.arange(1,ny-1)
        
        #w[ix_(i,j)] = (v[ix_(i+1,j)]-v[ix_(i-1,j)])/2./dx - (u[ix_(i,j+1)]-u[ix_(i,j-1)])/2./dy
        w[1:-1,1:-1] = (v[2:,1:-1]-v[0:-2,1:-1])/2./dx - (u[1:-1,2:]-u[1:-1,0:-2])/2./dy

    
        i = np.arange(1,nx-1); j = 0            #----- Bottom points
        w[i,j] = (self.uBCB-u[i,j+1])/dy

        i = np.arange(1,nx-1); j = ny-1         #----- Top points
        w[i,j] = (u[i,j-1]-self.uBCT)/dy

        i = 0; j = np.arange(1,ny-1)            #----- West points
        w[i,j] = (v[i+1,j]-self.vBCW)/dx

        i = nx-1; j = np.arange(1,ny-1)         #----- East points
        w[i,j] = (self.vBCE-v[i-1,j])/dx

        return
    
    #======================================================
    
    def rhsf(self) :
        """ Do the Psi solve for the current omega,
            Then compute omega boundary conditions,
            Then find the omega rate (RHS function).
        """
        
        u = self.u
        v = self.v
        w = self.w
        P = self.P
        dx = self.dx
        dy = self.dy
        nx = self.nx
        ny = self.ny
        nu = self.nu
        ix_ = np.ix_          # for convenience, so we can write ix_ instead of np.ix_
    
        i = np.arange(1,nx-1)
        j = np.arange(1,ny-1)
    
        self.setPfromW()   
        #self.setPfromW_adi()        

    
        #-------- set w boundaries
        # note, P on boundaries is zero, but included anyway
    
        w[0,1:-1]  = 2./dx/dx * (P[0,1:-1]  - P[1,1:-1]  - dx*self.vBCW)   # left
        w[-1,1:-1] = 2./dx/dx * (P[-1,1:-1] - P[-2,1:-1] + dx*self.vBCE)   # right
        w[1:-1,0]  = 2./dy/dy * (P[1:-1,0]  - P[1:-1,1]  + dy*self.uBCB)   # bottom
        w[1:-1,-1] = 2./dy/dy * (P[1:-1,-1] - P[1:-1,-2] - dy*self.uBCT)   # top
    
        #-------- get the rhs

        R = np.zeros((nx,ny))

        R[ix_(i,j)] = -1./4./dx/dy*(P[ix_(i,j+1)]-P[ix_(i,j-1)])*(w[ix_(i+1,j)]-w[ix_(i-1,j)])+1./4./dx/dy*(P[ix_(i+1,j)]-P[ix_(i-1,j)])*(w[ix_(i,j+1)]-w[ix_(i,j-1)]) + nu/dx/dx*(w[ix_(i-1,j)]-2.*w[ix_(i,j)]+w[ix_(i+1,j)]) + nu/dy/dy*(w[ix_(i,j-1)]-2.*w[ix_(i,j)]+w[ix_(i,j+1)])
        
        return R
    
    #======================================================
   
    def setUVfromP(self) :
        """ Set u, v from psi, for plotting """
    
        i = np.arange(1,self.nx-1)
        j = np.arange(1,self.ny-1)
        ix_ = np.ix_
    
        self.u[ix_(i,j)] =  (self.P[ix_(i,j+1)] - self.P[ix_(i,j-1)])/2.0/self.dy
        self.v[ix_(i,j)] = -(self.P[ix_(i+1,j)] - self.P[ix_(i-1,j)])/2.0/self.dx
    
        return
    
    #======================================================

    def setPfromW(self) :
        """ Set Psi from omega.  Called by the rhsf function. """
    
        nxm = self.nx-2
        nym = self.ny-2
    
        a = (-2.0/self.dx/self.dx - 2.0/self.dy/self.dy) * np.ones(nxm*nym)
        b = 1.0/self.dx/self.dx * np.ones(nxm*nym)
        c = 1.0/self.dy/self.dy * np.ones(nxm*nym) 
        d = b.copy()
        e = c.copy()  
    
        d[nxm:nxm*nym+1:nxm]    = 0.0     # spdiags chops first elem of d
        b[nxm-1:nxm*nym+1:nxm]  = 0.0     # spdiags chops last elem of b

        #A = np.diag(c[:-nxm],-nxm)+np.diag(b[:-1],-1)+np.diag(a,0)+np.diag(d[1:],1)+np.diag(e[:-nxm],nxm)
        A = scipy.sparse.diags((c[:-nxm],b[:-1],a,d[1:],e[:-nxm]),[-nxm,-1,0,1,nxm])
        # boundary values of P are zero, so don't fix B for BC's
             
        #p_interior = np.reshape(np.linalg.solve(A,-self.w[1:-1,1:-1].ravel()),(nxm,nym))
        p_interior = np.reshape(scipy.sparse.linalg.spsolve(A,-self.w[1:-1,1:-1].ravel()),(nxm,nym))

        self.P[1:self.nx-1, 1:self.ny-1] = p_interior
        
        return
    
    #======================================================
        
    def setPfromW_adi(self) :
        """ The adi version of the Psi solver.  """

        nx = self.nx
        ny = self.ny
        dx = self.dx
        dy = self.dy
        w  = self.w
        P  = self.P
        ix_= np.ix_
        
        maxit = 500
        tol   = 1.0E-4

        nxm = nx-2
        nym = ny-2

        a_x = -(2.0/dy/dy+2.0/dx/dx)*np.ones(nxm)
        d_x = np.ones(nxm)/dx/dx
        c_x = np.ones(nxm)/dx/dx

        a_y = -(2.0/dy/dy+2.0/dx/dx)*np.ones(nym)
        d_y = np.ones(nym)/dy/dy
        c_y = np.ones(nym)/dy/dy
        
        for it in range(maxit) :
            
            #------ sweep i lines  (soln in y dir)

            j = np.arange(1,ny-1)
            for i in range(1,nx-1) :
                b_y = -w[i,j] - 1.0/dx/dx*( P[i-1,j] + P[i+1,j] )
                P[i,j] = self.thomas(a_y,d_y,c_y,b_y)

            #------ sweep j lines  (soln in x dir)

            i = np.arange(1,nx-1)
            for j in range(1,ny-1) :
                b_x = -w[i,j] - 1.0/dy/dy*( P[i,j-1] + P[i,j+1] )
                P[i,j] = self.thomas(a_x,d_x,c_x,b_x)

            #------ check convergence

            i = np.arange(1,nx-1)
            j = np.arange(1,ny-1)
            res = (P[ix_(i-1,j)]-2.0*P[ix_(i,j)]+P[ix_(i+1,j)])/dx/dx + (P[ix_(i,j-1)]-2.0*P[ix_(i,j)]+P[ix_(i,j+1)])/dy/dy + w[ix_(i,j)]
            err = np.linalg.norm(res)
            ref = np.linalg.norm(w)
            if err/ref < tol :
                break
        

    #======================================================

    def thomas(self,a_in,d,c,b_in) :
        """ Called by the setPsiFromW_adi function. """
        
        a = a_in.copy()        # a is modified, but we are reusing a, so make a copy
        b = b_in.copy()        # b too

        n = np.size(a)
        x = np.zeros(n)
        for i in range(1,n) :
            a[i] = a[i] - c[i]/a[i-1]*d[i-1]
            b[i] = b[i] - c[i]/a[i-1]*b[i-1]
        
        x[n-1] = b[n-1]/a[n-1]
        for i in range(n-2,-1,-1) :
            x[i] = (b[i]-d[i]*x[i+1])/a[i]
        
        return x
```
