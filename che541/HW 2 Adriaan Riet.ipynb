<h1>#2</h1>
<p>Compute $(1+1/n)^n$ for $n=10^k$, $k=1,2,...,20$</p>

```python
def compE(k):
    n=10**k
    return (1+1/n)**n
cE = np.vectorize(compE)
x = np.linspace(1,20,20)
y1= compE(x)
y2= cE(x)
n = 10**x
error1 = np.abs(y1 - np.exp(1))/np.exp(1)
error2 = np.abs(y2 - np.exp(1))/np.exp(1)
plt.subplot(1,2,1)
plt.plot(x,cE(x),'b-',x,compE(x),'r-')
plt.xlabel('k')
plt.ylabel('e')
plt.subplot(1,2,2)
plt.semilogy(x,error1,'b-',x,error2,'r-')
plt.xlabel('k')
plt.ylabel('Relative Error')
plt.tight_layout()

print("""The error jumps up radically after n reaches about {:1.0g}, which makes sense, as we
      start reaching the limit of machine precision at that point""".format(10**x[14]))
```

<h1>#6</h1>
<p>Write a program implementing the Thomas algorithm and use it to solve equation (J) of Hoffman section 1.5</p>

```python
def thomasAlg(Alower,Amiddle,Aupper,b):
    n,=np.shape(Amiddle)
    for i in range(n-1):
        Alower[i]    = Alower[i]/Amiddle[i]
        Amiddle[i+1]-= Aupper[i]*Alower[i]
        b[i+1]      -= Alower[i]*b[i]
    x = b.copy()
    x[-1]/=Amiddle[-1]
    for i in range(n-2,-1,-1):
        x[i]-=Aupper[i]*x[i+1]
        x[i]/=Amiddle[i]
    return x
    
A1=np.array([2,2,2],dtype=np.float64)
A2=np.array([3,3,3,3],dtype=np.float64)
A3=np.array([2,2,2],dtype=np.float64)
b =np.array([12,17,14,7],dtype=np.float64)
x = thomasAlg(A1,A2,A3,b)
A = np.array([[3,2,0,0],[2,3,2,0],[0,2,3,2],[0,0,2,3]],dtype=np.float64)
b2= np.array([12,17,14,7],dtype=np.float64)
print(x)
print(np.dot(A,x)-b2)
x2=np.linalg.solve(A,b2)
print(x2)
```
