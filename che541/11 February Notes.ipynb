<h3>Stability:</h3>
<p>A finite difference equation is stable if it produces bounded solutions.</p>

```python
x = np.linspace(1e-6,15)
y = (x-15)**2 + 40/x*np.sin(5*x)
y2 = (x-15)**2
plt.subplot(1,2,1)
plt.plot(x,y,label='approximate')
plt.plot(x,y2,label='Exact')
plt.xlabel('stable')
plt.legend()
plt.subplot(1,2,2)
x = np.linspace(1e-6,15)
y = (x-15)**2 + 20*x*np.sin(5*x)
y2 = (x-15)**2
plt.plot(x,y,label='approximate')
plt.plot(x,y2,label='Exact')
plt.xlabel('unstable')
plt.legend()
```
