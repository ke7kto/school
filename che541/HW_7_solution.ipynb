### Solution

The first three parts of this problem are solved with the omega method:
$$\alpha T_{i-1}^{n+1} - (2\alpha+ 1/\Delta t)T_i^{n+1} +\alpha T_{i+1}^{n+1} = -S_i -\beta T^n_{i-1} + (2\beta - 1/\Delta t)T_i^n - \beta T^n_{i+1}.$$

Here, $\alpha = \chi\omega/2\Delta\xi^2$, and $\beta = \chi(1-\omega)/2\Delta\xi^2$. When $\omega = 0.0$, we get the explicit FTCS, method, when $\omega = 0.5$, we get the implicit Crank Nicholson method,
the implicit Crank Nicholson method, and when $\omega = 1$, we get the implicit BTCS method. The method of lines is solved an internal ODE solver routine with the RHS function discretized as shown in the code below. 

The first 4 figures below shows the evolution of the solution in time for the four methods. The FTCS method requires a small stepsize for stability, so the mesh plot appears filled in. A uniform stepsize of 8.333E−7 is used for the FTCS method, and a timestep corresponding to 10 timesteps is used for the CN and BTCS methods. The MOL stepsize varies and is specified by the solver defaults. The number of timesteps is set deliberately small for the implicit methods to illustrate the unconditional stability of these methods.

The final figure below shows the time evolution of the stoichiometric point (the peak flame temperature)
at a mixture fraction of $\xi = 0.2$. Note that the maximum timestep for stability of the FTCS method
is 1.0E−6, slightly larger than the value used. This is determined by the criteria $d\le 0.5$, where
$d = 0.5\chi\Delta t/\Delta\xi^2$. $\chi$ varies on the grid, and this has to be accounted for in determining the maximum stepsize. Note, however, that the MOL and the FTCS methods are very close, and accurate. Hence, the small stepsizes associated with stability, also give good accuracy here. In contrast, the BTCS method is stable but inaccurate for the few timesteps used. The CN method is a second order method, so it should be more accurate, but for the few timesteps used, the method gives a stable, but oscillatory solution. The oscillations center about the MOL solution better than the BTCS method. These oscillations decrease and dissappear as the stepsize is decreased.
The scalar dissipation rate $\chi$ represents flame strain. As this parameter increases, eventually the temperature source term cannot compete and the flame will extinguish. Here, we have taken $\chi$ and $S$ as constants, but in reality, they are temperature and composition dependent.
The code for the solution is given below.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
from scipy import integrate
from scipy import interpolate
```

```python
N    = 22                   # CHANGEME: total Grid pts w/ boundaries
chi0 = 2500.0               # chi scale: chi = cho*g(f)
n    = N-2                  # interior points
Tbc  = 298.15               # Boundary temperature

fa   = np.linspace(0,1,N)                  # all grid points f
f    = fa[1:-1]                            # interior f points
df   = f[1] - f[0]                         # grid spacing
chia = chi0*(1.0-(2.0*fa-1.0)**2.0)**2.0  # chi profile
chi  = chia[1:-1]                          # interior chi points

#--------------------- Set initial conditions and T source term

fTS = np.array([
[0.0,         Tbc,                 0.0],
[0.047619048,  837.9553388, 510.3061318],
[0.095238095, 1377.760678, 1269297.328],
[0.142857143, 1917.566016, 18207356.96],
[0.19047619 , 2457.371355, 30622770.11],
[0.238095238, 2590.523339, 23825333.14],
[0.285714286, 2447.250005, 14987151.69],
[0.333333333, 2303.976671, 8796653.546],
[0.380952381, 2160.703338, 4931526.203],
[0.428571429, 2017.430004, 2608687.109],
[0.476190476, 1874.15667 , 1273323.674],
[0.523809524, 1730.883337, 558386.2185],
[0.571428571, 1587.610003, 213038.1416],
[0.619047619, 1444.336669, 67832.97339],
[0.666666667, 1301.063336, 17015.25398],
[0.714285714, 1157.790002, 3090.090753],
[0.761904762, 1014.516668, 356.7225523],
[0.80952381 , 871.2433347, 21.17058059],
[0.857142857, 727.970001 , 0.44501952],
[0.904761905, 584.6966673, 0.00160563],
[0.952380952, 441.4233337, 0.00000019],
[1.0,         Tbc,         0.0]])

f_spline  = interpolate.interp1d(fTS[:,0],fTS[:,1], kind='cubic')
T0        = f_spline(f)
f_spline  = interpolate.interp1d(fTS[:,0],fTS[:,2], kind='cubic')
S         = f_spline(f)

#----------------- Now you have your f grid,
#----------------- and T0 is the initial condition on your grid
#----------------- and S on your grid

##################################################################
# Solve the ODE system
##################################################################


##################################################################
# Plot the results
##################################################################

```

```python
def f_omega(T, dt, omegafac) :
    
    aa = chi*omegafac/2.0/df/df
    bb = chi*(1.0-omegafac)/2.0/df/df
    a = -(2.0*aa+1.0/dt)
    c = aa
    
    A = np.diag(c[1:],-1) + np.diag(a,0) + np.diag(c[:-1],1)

    T = np.hstack((Tbc,T,Tbc))
    
    i = np.arange(1,N-1)
    b = -S - bb*T[i-1] + (2.0*bb-1.0/dt)*T[i] - bb*T[i+1]
    b[0] = b[0] -Tbc*aa[0]
    b[-1] = b[-1] - Tbc*aa[-1]
    
    return np.linalg.solve(A,b)
```

### FTCS

```python
t0 = 0
tf = 0.001

omegafac = 0.0
nt = 1200
dt = (tf-t0)/nt
t  = np.linspace(t0,tf,nt+1)
T = np.zeros((nt+1,n))
T[0,:] = T0

dtmax = np.min(df*df/chi)
ntmin = (tf-t0)/dtmax
if nt<ntmin :
    print "drop your timestep or else"

for i in range(nt) :
    T[i+1,:] = f_omega(T[i,:], dt, omegafac)
TfFTCS = T[-1,:]
TfFTCS = np.hstack((Tbc,TfFTCS,Tbc))

TTftcs = T[:,3]
tftcs = t

#-------------------- plot the results

ff,tt = np.meshgrid(f,t)

plt.rc("font", size=14)   
fig = plt.figure(figsize=(8,5))
ax  = fig.add_subplot(1,1,1, projection='3d')
p   = ax.plot_surface(ff, tt*1000.0, T, rstride=1, cstride=1, 
                      cmap=cm.jet, linewidth=0.0)
ax.xaxis._axinfo['label']['space_factor'] = 2
ax.yaxis._axinfo['label']['space_factor'] = 2
ax.zaxis._axinfo['label']['space_factor'] = 1.8
ax.set_xlabel(r'$\xi$', size=16)
ax.set_ylabel('Time (ms)')
ax.set_zlabel('T (K)', rotation=180)
ax.set_title('FTCS')
ax.view_init(35,300)

```

### Method of Lines

```python
def rhsf(T,t) :
    
    rr = np.zeros(n)
    
    i=0      # node next to lower boundary
    rr[i] = 0.5*chi[i]/df/df*( Tbc-2.0*T[i]+T[i+1] ) + S[i]

    i=np.arange(1,n-1)  # internal nodes
    rr[i] = 0.5*chi[i]/df/df*( T[i-1]-2.0*T[i]+T[i+1] ) + S[i]

    i=n-1    # node next to upper boundary
    rr[i] = 0.5*chi[i]/df/df*( T[i-1]-2.0*T[i]+Tbc ) + S[i]
    
    return rr
```

```python
times_to_solve = np.linspace(0,0.001,20)

T = integrate.odeint(rhsf,T0,times_to_solve)

TfMOL = T[-1,:]
TfMOL = np.hstack((Tbc,TfMOL,Tbc))

TTmol = T[:,3]
tmol = times_to_solve

#-------------------- plot the results

ff,tt = np.meshgrid(f,times_to_solve)

plt.rc("font", size=14)   
fig = plt.figure(figsize=(8,5))
ax  = fig.add_subplot(1,1,1, projection='3d')
p   = ax.plot_surface(ff, tt*1000.0, T, rstride=1, cstride=1, 
                      cmap=cm.jet, linewidth=1.0)
ax.xaxis._axinfo['label']['space_factor'] = 2
ax.yaxis._axinfo['label']['space_factor'] = 2
ax.zaxis._axinfo['label']['space_factor'] = 1.8
ax.set_xlabel(r'$\xi$', size=16)
ax.set_ylabel('Time (ms)')
ax.set_zlabel('T (K)', rotation=180)
ax.set_title('MOL')
ax.view_init(35,300)
```
