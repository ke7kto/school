```python
x = np.linspace(-10,10,100)
y = x**2
y2 = x**2 + 30*np.sin(x*10/(4*pi))
miny = np.argmin(y2)
sminy = np.argmin(y2[50:])+50
zminy = np.argmin(y2[:20])
with plt.xkcd():
    fig = plt.figure(frameon=False)
    plt.plot(x,y)
    plt.xlim(-10,10)
    plt.ylim(-10,100)
    plt.xlabel(r'$\vec{\alpha}$',fontsize=18)
    ax = fig.gca()
    plt.xticks([])
    plt.yticks([])
    plt.ylabel('Sum',fontsize=18)
    plt.annotate(r'Best $\vec{\alpha}$', xy=(0.5,-0.5),arrowprops=dict(arrowstyle='-|>'),xytext=(5,-5))
    #plt.savefig('singleMinimum.png')
    fig2 = plt.figure()
    plt.plot(x,y2)
    ax = fig.gca()
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(r'$\vec{\alpha}$',fontsize=18)
    ax = fig.gca()
    plt.xticks([])
    plt.yticks([])
    plt.ylabel('Sum',fontsize=18)
    plt.annotate(r'Best $\vec{\alpha}$', xy=(x[miny]+0.5,y2[miny]-0.5),arrowprops=dict(arrowstyle='-|>'),xytext=(5,-15))
    plt.annotate(r'Local minima, can cause problems', xy=(x[sminy]+0.5,y2[sminy]-0.5),arrowprops=dict(arrowstyle='-|>'),xytext=(-5.5,80))
    plt.annotate(r'Local minima, can cause problems', xy=(x[zminy]+0.5,y2[zminy]-0.5),arrowprops=dict(arrowstyle='-|>'),xytext=(-5.5,80))
    #plt.savefig('multipleMinima.png')
```

```python
y = 2*(np.random.random(np.size(x))-0.5)
y2 = y + 1.5*np.cos(x)
fig = plt.figure(figsize=(14,2))
fig.add_subplot(131)
plt.title('Good')
plt.plot(x,y,'b.')
plt.xticks([])
plt.yticks([0])
plt.ylim(-10,10)
fig.add_subplot(132)
plt.title('Bad')
plt.plot(x,y2,'b.')
plt.ylim(-10,10)
plt.xticks([])
plt.yticks([0])
fig.add_subplot(133)
plt.title('Bad')
plt.plot(x,y+1,'b.')
plt.ylim(-10,10)
plt.xticks([])
plt.yticks([0])
#plt.savefig('residuals.png',bbox_inches='tight')
```

```python
import sympy as s
s.init_printing()
theta = s.Symbol(r'\theta')
k = s.Symbol('k',positive=True)
vec = k*s.Matrix([-sin(theta),0,1-cos(theta)])
s.sqrt(vec.dot(vec)).simplify()
```

```python


mean = 0
variance = 1
sigma = np.sqrt(variance)
x = np.linspace(-3,3,1000)
fig = plt.figure()
plt.plot(x,scs.norm.pdf(x,mean,sigma))
plt.savefig('gaussian.png')
fig = plt.figure()
plt.plot(x,scs.norm.pdf(x,mean,sigma/5))
plt.savefig('sharpgauss.png')
fig = plt.figure()
plt.plot(x,scs.norm.pdf(x,mean,sigma/25))
plt.savefig('verysharpgauss.png')
```
