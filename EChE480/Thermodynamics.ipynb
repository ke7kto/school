highly reversible: Rates are very fast. 

$$\Delta G = -nFE$$

Nernst Equation
$$E_\text{red} = E^0_\text{red} - \frac{RT}{nF}\ln\left(\frac{[\text{Reduced}]}{[\text{Oxidized}]}\right)$$

Reaction

$$\text{Ox} + ne^- \rightleftharpoons \text{Red}$$
Reduction Potential:
$$\Delta G = -nFE$$
Standard Reduction Potential:
$$\Delta G^0 = -nFE^0$$
Derive Nernst:
$$\Delta G = \Delta G^0 + RT\ln\left(\frac{a_\text{red}}{a_\text{ox}}\right)$$
$$-nFE = -nFE^0 +  RT\ln\left(\frac{a_\text{red}}{a_\text{ox}}\right)$$
$$ E = E^0 - \frac{RT}{nF}\ln\left(\frac{a_\text{red}}{a_\text{ox}}\right)\approx E^0 - \frac{RT}{nF}\ln\left(\frac{[\text{red}]}{[\text{ox}]}\right) $$
Where $E = E_\text{red}$ is known as the reduction potential, or the thermodynamic potential

Activity of metallic reduced species is equal to 1.

```python
import scipy.constants as c
```

```python
Fe = np.linspace(0,1,1000) #Molar, mol/L
E = -0.440 - c.R*298/(96485*2)*np.log(Fe)
Fe2 = np.logspace(-9,0,num=100,base=10)
E2 = -0.440 - c.R*298/(96485*2)*np.log(Fe2)
```

```python
fig = plt.figure(figsize=(18,6))
ax2 = fig.add_subplot(121)
plt.plot(Fe,E)
plt.xlabel('Concentration of Iron, (mol/L)')
plt.ylabel('Voltage (V)')
plt.ylim(-0.45,-0.15)
ax3 = fig.add_subplot(122)
plt.semilogx(Fe2,E2)
plt.ylim(-0.45,-0.15)
plt.xlabel('Concentration of Iron, (mol/L)')
plt.ylabel('Voltage (V)')
```
