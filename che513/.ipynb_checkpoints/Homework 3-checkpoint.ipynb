```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def shm(t,w=1,alpha=2,delta=3):
    x  =  np.cos(w*t+delta)
    px = -np.sin(w*t+delta)*w
    y  =  np.cos(w*t+alpha)
    py = -np.sin(w*t+alpha)*w 
    return x,y,px,py
t = np.linspace(0,2*np.pi)
x,y,px,py = shm(t,w=1,alpha=2,delta=2)
plt.plot(x,y,label='$x-y$',ls='-.')
plt.plot(px,x,label='$p_x - x$')
plt.plot(px,py,label='$p_x - p_y$',ls='-',lw=0.5)
plt.legend(loc='best')
g=plt.title('Part a')
```

```python
plt.figure()
x,y,px,py = shm(t,w=1,alpha=2,delta=2+np.pi/2)
plt.plot(x,y,label='$x-y$',ls=':',lw=5)
plt.plot(px,x,label='$p_x - x$',ls='--',lw=3)
plt.plot(px,py,label='$p_x - p_y$',ls='-',lw=1)
plt.legend(loc='center')
g=plt.title('Part b')
```

```python
plt.figure()
x,y,px,py = shm(t,w=1,alpha=2,delta=2-np.pi/4)
plt.plot(x,y,label='$x-y$',ls=':')
plt.plot(px,x,label='$p_x - x$')
plt.plot(px,py,label='$p_x - p_y$',ls='-',lw=0.5)
plt.legend(loc='best')
plt.title('Part c')
print("It is interesting how the phase space changes as we change the phase of x and y relative to each other")
```

<h3>Problem 3</h3>
<p>The partition function of an ideal gas of diatomic molecules in an external electric field is<br/>
$Q(N,V,T,E) = \frac{\left[q(V,T,E)\right]^N}{N!}$<br/>
Where<br/><ul><li>$q(V,T,E) = V\Lambda^3 \cdot q_r\cdot q_v\cdot \left(\frac{kT}{\mu E}\right)\sinh\left(\frac{\mu E}{kT}\right)$</li>
<li>$q_r = \frac{8\pi^2 IkT}{h^2}$</li>
<li>$q_v = \frac{\exp\left(\frac{-h\nu}{2kT}\right)}{1-\exp\left(\frac{-h\nu}{kT}\right)}$</li>
</ul>
$\def\D{\mathrm{d}}$
I is the moment of inertia of the molecule, $\nu$ is the molecule's fundamental vibrational
frequency, and $\mu$ is the molecule's permanent dipole moment. Using this partition function
and the thermodynamic relation <br/>$\D{A} = -S\D{T} – P\D{V}– N\langle\mu\rangle\D{E}$<br/>
show that<br/>
$\langle \mu \rangle = \mu\left[\coth\left(\beta\mu E\right) - \frac{1}{\beta\mu E}\right]$<br/>
where $\langle\mu\rangle$ is the average dipole moment of a molecule in the direction of the external field.
Then sketch the functionality of the dimensionless dipole moment $\mu^+ = \frac{\langle \mu \rangle}{\mu}$ result versus
the dimensionless field $E^+$ $\left(E^+ = \beta\mu E\right)$ from $E^+ = 0$ to $E^+ = \infty$ and interpret what you see.
<details><ol>
<li>The "E" in the problem statement is not energy, it is electric field</li>
<li>The following relationships need to be used. <br/>
$\left(\frac{\partial A}{\partial E} \right)_{T,V} = -N\mu$<br/>
and<br/>
$A = -kT\ln{Q}$</li></ol>

<h5>Answer</h5>
<ol><li>$\D{A} = -S\D{T} – P\D{V}– N\langle\mu\rangle\D{E}$<br/>
$\frac{\D{A}}{\D{E}} = - N\langle\mu\rangle = \frac{\D{\left(-kT\ln Q\right)}}{\D{E}}$
</li>
<li>$\ln Q = \ln\left(\frac{\left[q(V,T,E)\right]^N}{N!}\right) = N \ln\left(q(V,T,E)\right) - \left(N\ln N - N\right)$</li>
<li>$\ln q = \ln\left(V\Lambda^3\right)+\ln\left(\cdot q_r\right)+\ln\left( q_v\right)+\ln\left(\frac{kT}{\mu E}\right)+\ln\left(\sinh\left(\frac{\mu E}{kT}\right)\right)$<br/>
$\ln q = \ln\left(V\Lambda^3\right)+\ln\left( \frac{8\pi^2 IkT}{h^2}\right)+\ln\left( \frac{\exp\left(\frac{-h\nu}{2kT}\right)}{1-\exp\left(\frac{-h\nu}{kT}\right)}\right)+\ln\left(\frac{kT}{\mu E}\right)+\ln\left(\sinh\left(\frac{\mu E}{kT}\right)\right)$
</li>
<li>$\frac{\D{\left(-kT\ln Q\right)}}{\D{E}} = -kT\left[N \ln\left(q(V,T,E)\right) - \left(N\ln N - N\right)\right]$
$=-kTN\frac{\D{\left[\ln\left(\frac{1}{ E}\right)+\ln\left(\sinh\left(\frac{\mu E}{kT}\right)\right)\right]}}{\D{E}}$
$=-kTN\frac{\D{\left[-\ln\left( E\right)+\ln\left(\sinh\left(\frac{\mu E}{kT}\right)\right)\right]}}{\D{E}}$
$=-kTN\left(\frac{-1}{E} +\frac{\D{\left[\ln\left(\sinh\left(\frac{\mu E}{kT}\right)\right)\right]}}{\D{E}}\right)$

<h3>Problem 4</h3>
<p>A rigid diatomic molecule rotates freely in three dimensions in a zero potential-energy field. The Hamiltonian may be expressed in terms of the moment of inertia as <br/>
$H = \frac{1}{2I}\left(p_\theta^2 + \frac{p_\phi^2}{\sin^2 \theta}\right)$<br/>
where $\theta$ ranges from $0$ to $\pi$, $\phi$ ranges from $0$ to $2\pi$ and <br/>
$p_\theta = I\dot{\theta}$, $p_\phi = I\dot{\phi}\sin^2\theta$<br/>
Evaluate the classical molecular partition function, $Q_{cl}$ (excluding the $N!$ term). Note that there are only two applicable coordinates, $\theta$ and $\phi$, because the distance between the atoms is fixed and translational motion is not considered.

$Q_{cl} = \frac{1}{N!h^{3N}} \int\limits_\tau e^{-\beta H}\D{\tau}=\int\limits_\phi\int\limits_\theta e^{-\beta H}\D{\theta}\D{\phi}=\int\limits_0^{2\pi}\int\limits_0^{\pi} e^{-\beta H}\D{\theta}\D{\phi}$
$=\int\limits_0^{2\pi}\int\limits_0^{\pi}\int\limits_{-\infty}^\infty \int\limits_{-\infty}^\infty e^{-\beta \frac{1}{2I}\left(p_\theta^2+\frac{p_\phi^2}{\sin^2 \theta}\right)}\D{p_{\theta}}\D{p_\phi}\D{\theta}\D{\phi}$<br/>


$=\int\limits_0^{\pi}\int\limits_{-\infty}^\infty \int\limits_{-\infty}^\infty e^{-\beta \frac{1}{2I}\left(p_\theta^2+\frac{p_\phi^2}{\sin^2 \theta}\right)}\D{p_{\theta}}\D{p_\phi}\D{\theta}    \int\limits_0^{2\pi}\D{\phi}$
$=\int\limits_0^{\pi} \int\limits_{-\infty}^\infty e^{-\beta \frac{1}{2I}\left(\frac{p_\phi^2}{\sin^2 \theta}\right)}\D{p_\phi}\D{\theta}\int\limits_{-\infty}^\infty e^{-\beta \frac{1}{2I}\left(p_\theta^2\right)}   \D{p_{\theta}}\int\limits_0^{2\pi}\D{\phi}$<br/>
$\int\limits_{-\infty}^\infty e^{-\frac{2}{3}x^2}\D{x} = \sqrt{\frac{3\pi}{2}}$<br/><br/>
$Q_{cl}=\int\limits_0^{\pi} \int\limits_{-\infty}^\infty e^{-\beta \frac{1}{2I}\left(\frac{p_\phi^2}{\sin^2 \theta}\right)}\D{p_\phi}\D{\theta}\sqrt{\frac{2I\pi}{\beta}}2\pi$
$=\int\limits_0^\pi \sqrt{\pi 2I kT}\sin(\theta)\D{\theta}$
$=\frac{2\sqrt{2\pi I kT}}{h^{3N}N!}$
